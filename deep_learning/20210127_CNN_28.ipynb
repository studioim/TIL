# 28. Ch 06. Convolutional Neural Networks - 01. CNN 소개

```python
<딥러닝 이전>
-Hand-crafted feature(예. edge 가장자리) 추출 후 모델링
내가 원하는 피처를 추출하기 위해서 특히 이미지(컴퓨터 비전) 분야에서는 컨볼루션 필터를 많이 사용했다.
딥러닝에서는 컨볼루션 레이어를 통과하면서 자동으로 최적의 피처를 추출 
컨볼루션 필터는 x와 y의 관계를 찾는 과정에서 정보 추출하기 위한 최적의 필요한 패턴들을 찾아내는 필터로 자동적으로 구성된다.
그게 바로 딥러닝에서의 컨볼루션 레이어의 특징.

<Convolution Operation>
-Convolution filter가 입력(x)에 대해서 돌아가며 동작
6 X 6 입력 데이터가 3 X 3 컨볼루션 필터를 통과하면 4 X 4 출력 데이터가 된다.

<Convolutional Neural Network>
input(4*4) X kernel(3*3) = output(2*2)
커널과 같은 위치에 있는 인풋 데이터를 곱해서 모두 더한 값들이 모여 아웃풋이 된다.

리니어 레이어의 경우 웨이트가 n X m일 때 n차원의 입력을 받아서 m 차원의 출력을 뱉어줬다.
지금은 커널이 돌아다니면서 동작하기 때문에 인풋의 사이즈나 아웃풋의 사이즈가 정의돼 있지 않다.
커널이 주어졌을 때 인풋의 사이즈에 따라서 아웃풋의 사이즈도 달라지는 구조.

가장자리 부분에는 커널이 들어갈 수 없다.
이를 보완한 것이 패딩

<CNN(with Pad)>
인풋 주변을 0으로 가득찬 패드로 둘러놓는 것.
패딩 한 칸이라고 하면 좌우로 한 칸 씩 둘러놓는다는 것.
두 칸 두를 수도 있다. 원하는 만큼 가능.
(4*4 + 1)X(3*3) = (4*4)
패딩을 통해서 입력과 출력의 크기를 맞춰줄 수 있다.

<How it works(intuitively 직관적으로)>
인풋 데이터가 필터와 일정(예.두 개) 이상 일치하면 히트맵(아웃풋)에 불이 들어온다.
특징이 특정한 형태를 띄게 된다면 경험상 2였다는 것을 뉴럴 네트워크가 기억하고 있다면 2를 출력해낸다.

<인풋 & 아웃풋 채널>
#인풋 채널s(4) : 예. 컬러 이미지(레드, 그린, 블루)
X
#아웃풋 채널s(3) = #kernels
=
# kernels(3)
 
<인풋 & 아웃풋 텐서s>
(N, C, H, W) 배치 사이즈/채널의 개수(R,G,B)/높이/너비

<입출력 크기 계산 방법>
컨볼루션 레이어는 FCL과는 달리 입출력의 형태가 정해져 있지 않다.
특히 파이토치는 인터프리터처럼 그때그때마다 연산이 수행. 케라스나 텐서플로처럼 컴파일되는 게 아님.
입력출력 텐서의 크기가 완전히 가변이어도 된다.
커널의 사이즈만 고정돼 있으면 됨.
커널 사이즈가 고정돼 있을 때면 입력 사이즈에 따라 출력 사이즈가 바뀜.
커널 사이즈가 변하면 입력 크기가 정해져 있어도 출력 사이즈 변한다.
(xheight - kheight + 1, xwidth - kwidth + 1)

<입출력 크기 계산 방법(+패드)>
(xheight + 2 * pheight - kheight  + 1, xwidth + 2 * pwidth - kwidth + 1)

3 x 3 커널이 제일 많이 쓰인다. 3 x 3 커널에 패딩을 하나 써서 입출력 사이즈가 보존되는 방법을 제일 많이 씀.
계속해서 출력이 입력과 똑같은 사이즈가 유지되니 차원 축소가 필요한 것.
그때 사용하는 것이 맥스 풀링, 스트라이드.

<컨볼루션 레이어의 특징>
-피처의 위치에 구애받지 않는다.
FCL 같은 경우(리니어 레이어)에는 어떤 피처가 어디에 위치해 있는가가 중요.
컨볼루션 레이어는 피처가 어디든지 나타날 수 있다.
FCL의 경우에는 엠니스트 이미지의 위치가 달라지면 아예 다른 데이터가 된다.
그래서 컨볼루션 레이어는 이미지에 특화.

<컨볼루션 레이어의 장점>
-같은 입출력을 갖는 FC레이어에 비해 더 적은 웨이트를 가진다.

 -병렬 계산 구성이 쉬우므로, GPU에서의 연산이 매우 빠르다.
필터가 움직이면서 진행되는 연산이 서로 의존도가 전혀 없다.
FCL은 CPU, GPU 돌릴 때 큰 차이가 없다. 심지어 CPU가 빠를 때도 있음.
컨볼루션 레이어는 GPU에서 엄청난 빠른 속도를 보여줌.

<컨볼루션 레이어의 단점>
-FC 레이어에 비해 입출력 크기가 계산이 까다로워, 네트워크 구성이 쉽지 않다.

<요약>
-예전부터 영상 처리 분야에서는 컨볼루션 필터를 활용
다만 탐지하고자 하는 패턴에 따라 필터를 손으로 하나하나 깎아서 사용했다.
딥러닝은 컨볼루션 레이어를 쓰는 순간 자동으로 커널을 찾아낸다.
-CNN은 데이터셋의 x와 y의 구성에 따라 자동으로 탐지해야 할 패턴을 추출
따라서 내부 커널은 해당 패턴을 추출하기 위한 형태로 자동으로 구성됨
-FC레이어에 비해 매우 빠르고 적은 웨이트 파라미터를 가짐.
-이미지에 굉장히 특화된 특성을 가지고 있음.
```
