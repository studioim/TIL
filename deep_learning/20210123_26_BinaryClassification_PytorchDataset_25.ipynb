# Binary Classification

## Load Dataset from sklearn

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
df['class'] = cancer.target

df.tail()
```

```python
df.columns
```

## Convert to Pytorch Tensor

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
data = torch.from_numpy(df.values).float()
# 데이터프레임을 텐서로 불러오기, 그냥 불러오면 더블이 되는데 이걸 확실히 해주기 위해 플롯으로 형변환

data.shape
```

```python
x = data[:, :10]
y = data[:, -1:]

print(x.shape, y.shape)
```

```python
# Train / Valid / Test ratio

ratios = [.6, .2, .2]
```

```python
train_cnt = int(data.size(0) * ratios[0])
valid_cnt = int(data.size(0) * ratios[1])
test_cnt = data.size(0) - train_cnt - valid_cnt
cnts = [train_cnt, valid_cnt, test_cnt]

print("Train %d / Valid %d / Test %d samples." % (train_cnt, valid_cnt, test_cnt))
```

```python
indices = torch.randperm(data.size(0)) # random permutation(순열, 치환), 0부터 568까지의 무작위 수열이 나온다

x = torch.index_select(x, dim=0, index=indices)
y = torch.index_select(y, dim=0, index=indices)

x = x.split(cnts, dim=0)
y = y.split(cnts, dim=0)

for x_i, y_i in zip(x,y):
    print(x_i.size(), y_i.size())
```

- You don't need to shuffle them, if you have fixed train/valid/test set.
- MNIST는 트레인 셋 6만장 테스트 셋 1만장으로 고정돼 있다. 이때 셔플해줄 필요 없음. 트레인 셋만 트레인 밸리드 셋으로 랜덤 셔플해주면 됨.

## Set Hyper Parameters

```python
n_epochs = 10000 # 최대 1만번 수행
batch_size = 128 # 커스텀 데이터로더 만들기 전에 지정돼 있어야 함
print_interval = 500
early_stop = 100 # 100번 동안 밸리데이션 로스 계산이 없으면 끝낼 것.
```

## Get DataLoaders

```python
from torch.utils.data import Dataset, DataLoader
```

```python
class CustomDataset(Dataset):
    def __init__(self, data, labels):
        self.data = data
        self.labels = labels
        
        super().__init__()
    
    def __len__(self):
        return len(self.data)
    
    def __getitem__(self, idx):
        return self.data[idx], self.labels[idx]
```

```python
train_loader = DataLoader(
    dataset=CustomDataset(x[0], y[0]),
    batch_size=batch_size,
    shuffle=True, # Allow shuffling only for training set
)
# 트레인 로더는 무조건 셔플해줘야 한다

valid_loader = DataLoader(
    dataset=CustomDataset(x[1], y[1]),
    batch_size=batch_size,
    shuffle=False,
)
# 검증 셋은 셔플 False 해주는 것을 추천
# 랜덤하게 100개보다 같은 100개에 대해 테스트를 해주는 게 좋을 수 있으니

test_loader = DataLoader(
    dataset=CustomDataset(x[2], y[2]),
    batch_size=batch_size,
    shuffle=False,
)

print("Train %d / Valid %d / Test %d samples." % (
    len(train_loader.dataset),
    len(valid_loader.dataset),
    len(test_loader.dataset),
))
```

## Build Model & Optimizer

```python
model = nn.Sequential(
    nn.Linear(x[0].size(-1), 6),
    nn.LeakyReLU(),
    nn.Linear(6, 5),
    nn.LeakyReLU(),
    nn.Linear(5, 4),
    nn.LeakyReLU(),
    nn.Linear(4, 3),
    nn.LeakyReLU(),
    nn.Linear(3, y[0].size(-1)),
    nn.Sigmoid(), # Binary Classification을 하기 위해 소프트맥스 대신 시그모이드 함수 사용
)
# batch normalization도 안 썼다. 그래도 잘 되나 보자

model
```

```python
optimizer = optim.Adam(model.parameters()) # 아담 옵티마이저로 모델 최적화
```

## Train

```python
from copy import deepcopy

lowest_loss = np.inf
best_model = None

lowest_epoch = np.inf
```

```python
#from tqdm import tqdm
train_history, valid_history = [], []

for i in range(n_epochs):
    model.train()
    
    # We can reomove following code block. 데이터로더가 이 역할을 해주기 때문에 필요 없음.
    # ---------------------------------------------------
    # indices = torch.randperm(x[0].size(0))
    # x_ = torch.index_select(x[0], dim=0, index=indices)
    # y_ = torch.index_select(y[0], dim=0, index=indices)
    #
    # x_ = x_.split(batch_size, dim=0)
    # y_ = y_.split(batch_size, dim=0)
    # ---------------------------------------------------
    
    train_loss, valid_loss = 0, 0
    y_hat = []
    
    for x_i, y_i in train_loader:
        y_hat_i = model(x_i) # |x_i| = (bs, 10), |y_hat_i| = (bs, 1) 
        loss = F.binary_cross_entropy(y_hat_i, y_i)
        
        optimizer.zero_grad()
        loss.backward()
        
        optimizer.step()
        train_loss += float(loss) # This is very important to prevent memory leak.
        
    train_loss = train_loss / len(train_loader) # 평균 로스
    # 한 epoch의 트레이닝이 끝남
    
    # validation
    model.eval()
    with torch.no_grad():
        valid_loss = 0
        
        for x_i, y_i in valid_loader:
            y_hat_i = model(x_i)
            loss = F.binary_cross_entropy(y_hat_i, y_i)
            
            # optimzer, backward 없음
            
            valid_loss += float(loss)
            
            y_hat += [y_hat_i]
            
    valid_loss = valid_loss / len(valid_loader)
    
    train_history += [train_loss]
    valid_history += [valid_loss]
    # 한 epoch의 학습이 모두 끝남
    
    if (i + 1) % print_interval == 0:
        print('Epoch %d: train_loss=%.4e, valid_loss=%.4e, lowest_loss=%.4e' % (
            i + 1,
            train_loss,
            valid_loss,
            lowest_loss,
        ))
        
    if valid_loss <= lowest_loss:
        lowest_loss = valid_loss
        lowest_epoch = i
        
        best_model = deepcopy(model.state_dict())
    else:
        if early_stop > 0 and lowest_epoch + early_stop < i + 1:
            print("There is no improvemoent during last %d epochs." % early_stop)
            break

print("The best validation loss from epoch %d: %.4e" % (lowest_epoch + 1, lowest_loss))
# best model을 다시 리스토어
model.load_state_dict(best_model)
```

## Loss History

```python
# 파이썬 경고 무시
import warnings
warnings.filterwarnings(action='ignore')
```

```python
plot_from = 2

plt.figure(figsize=(20, 10))
plt.grid(True)
plt.title("Train / Valids Loss History")
plt.plot(
    range(plot_from, len(train_history)), train_history[plot_from:],
    range(plot_from, len(valid_history)), valid_history[plot_from:],
)
plt.yscale('log')
plt.show()
```

## Let's see the result!

```python
test_loss = 0
y_hat = []

model.eval()
with torch.no_grad():
    for x_i, y_i in test_loader:
        y_hat_i = model(x_i)
        loss = F.binary_cross_entropy(y_hat_i, y_i)
        
        test_loss += loss # Gradient is already detached.
        
        y_hat += [y_hat_i]
        
test_loss = test_loss / len(test_loader)
y_hat = torch.cat(y_hat, dim=0) # torch.cat은 y_hat에 있는 값을 모두 합쳐준다. 하나의 텐서로 만들기

print("Test_loss: %.4e" % test_loss)
```

```python
correct_cnt = (y[2] == (y_hat > .5)).sum() # 0.5보다 크면 1, 작으면 0
total_cnt = float(y[2].size(0)) # 나누기 위해 미리 float을 씌워줌

print("Test Accuracy: %.4f" % (correct_cnt / total_cnt))
```
