## 01. Ch 01. 추천시스템이란 - 01. 추천시스템, 그것이 궁금하다

```python
<추천 시스템이란?>
사용자(유저)와 상품(아이템)으로 구성된 시스템 
특정 사용자가 좋아할 상품을 추천
비슷한 상품을 좋아할 사용자를 추천
아이템이든 유저든 관심 갖을만한 정보를 추천

인터넷에서 정보를 찾는 데 시간이 매우 오래 걸림
사용자가 정보를 수집하고 찾는 시간을 줄여주는 것이 목적!
검색 서비스 / 추천 서비스
Pull Information / Push Information
사용자가 요구한 뒤 작동 / 사용자가 요구하기 전 작동
사용자 스스로 원하는 바를 알고 있다 / 사용자 스스로 원하는 바를 정확히 알지 못한다

<사용자와 상품>
사용자와 아이템 사이의 관계를 분석하고 연관관계를 찾는다
해당 연관 관계 점수화한다
사용자의 정보와 아이템의 정보를 활용한다

-사용자 정보
사용자 고유 정보(나이 성별 지역 등)
사용자 로그 분석(행동 패턴 등)
-아이템 정보
아이템 고유 정보(가격 색상 내용 등)

-사용자 프로필
사용자 또는 사용자 그룹을 분석 가능한 요소로 프로파일링
사용자를 구분할 수 있는 정보를 활용
	-사용자 아이디: 나이 성별 지역 학력 등 개인 신상정보
	-쿠키
	-인터넷 주소
	-웹페이지 방문 기록, 클릭 패턴 등 사용자 행동 정보
사용자 정보를 수집하기 위한 방법
	-직접적(Explicit) 방법: 설문조사, 평가, 피드백 등(회원 가입 등을 통해)
	-간접적(Implicit) 방법: 웹페이지 머무는 시간, 클릭 패턴, 검색 로그 등
개인별 추천 또는 사용자 그룹별 추천 가능

-아이템 프로필
플랫폼마다 정의하는 아이템의 종류가 다르다
일반적으로 생각해볼 수 있는 추천 아이템 예시
	-책, 전자 제품, 옷 등 웹사이트 내 웹페이지
	-SNS에서 비슷한 게시글, 관심 있을 만한 사진
	-뉴스 논문 등 문서
	-여행지 음식점 등 지역 또는 장소 정보
	-영화 음악 동영상 티비 프로그램 등 다양한 영상
모든 것을 추천해주는 플랫폼은 현재 없다
플랫폼마다 관련 있는 상품 또는 아이템만 추천한다
아이템 프로필에 속하는 정보
	-아이템 아이디, 아이템 고유 정보(크기 색 가격 등), 아이템을 좋아하거나 구매한 사용자 정보 등

<추천 점수란?>
분석된 사용자와 아이템 정보를 바탕으로 추천점수 계산
사용자 또는 아이템 프로필에서 어떤 정보를 사용할지에 따라 추천 알고리즘 결정
사용자 또는 아이템을 추천하기 위해 각각의 아이템 또는 사용자에 대한 정량화된 기준 필요
추천 알고리즘의 목적은 점수화(scoring)하는 것이다.

<이번 강의 리뷰>
추천 시스템은 아이템 또는 사용자를 추천한다
아이템과 사용자 프로필은 추천 알고리즘을 위한 피처가 된다
추천 알고리즘은 아이템 또는 사용자를 점수화하는 것이다

<다음 강의 프리뷰>
추천 시스템은 어디에 어떻게 사용될까?
추천 시스템을 만들기 위해 어떤 알고리즘이 있을까?
```

## 02. Ch 01. 추천시스템이란 - 02. 현재 추천시스템은

```python
<추천 시스템 베이직 리뷰>
추천시스템의 등장
수많은 정보로 사용자가 적절한 결정을 내리기 어려움

추천시스템의 정의
사용자로부터 선호도(preferences) 여부를 데이터화
	-Explicit Feedback(평점 등)
	-Implicit Feedback(사용자의 행동 패턴 등)
사용자의 선호도와 제한사항(constraints)을 바탕으로 가장 적절한 아이템 점수 계산
사용자에게 가장 관련 있는 아이템 순서대로 리스트 형태로 제시

추천 시스템은 검색 서비스에서 출발
정보 검색 분야(Information Retrieval)와 비슷하지만 비교적 새로운 연구 분야
넷플릭스 프라이즈 추천 대회 이후 많은 관심 불러

<추천 시스템을 왜 사용해야 할까?>
서비스 제공자는 추천시스템을 사용하고 싶은 다양한 이유가 있다
1.더 많은 아이템을 판매할 수 있다
상업 기업에서 추천 시스템을 사용하는 가장 큰 이유
2. 더 다양한 아이템을 판매할 수 있다
소비자가 보지 못한 상품을 판매할 수 있다.
3.소비자 만족도가 증가한다
플랫폼을 사용하면서 만족도 증가, 플랫폼에 더 머무르는 이유가 된다
4.충성도 높은 고객이 증가한다
5.고객이 원하는 것이 무엇인지 알 수 있다(니즈 파악)
데이터가 쌓이면 쌓일수록 추천 알고리즘의 성능 향상

<추천시스템은 어디에 사용될 수 있을까?>
-넷플릭스: 콘텐츠 추천
계정 생성 시 선택된 좋아하는 컨텐츠를 기반으로 콘텐츠 추천
-인스타그램: 팔로우 추천
팔로잉하고 있는 계정 유형과 피드에서 자주 클릭한 계정 분석해 피드에 광고 노출
계정 팔로우할 경우 해당 계정과 비슷한 계정 추가적으로 추천(푸쉬 인포메이션이 가능)
-쿠팡: 상품 추천
구매 목록, 상품 검색 키워드, 클릭한 상품을 기반으로 다양한 상품 추천

<추천 시스템: 데이터 종류>
아이템, 유저, Transactions

<추천시스템이 풀고자 하는 문제>
1.랭킹 문제
특정 유저가 특정 아이템에 대한 평점(또는 점수)을 정확하게 예측할 필요 없음
특정 아이템을 좋아할만한 Top-K 유저를 선정할 수 있음
특정 유저가 좋아할 만한 Top-K 아이템을 선정할 수 있음
2.예측 문제
유저-아이템 조합에서 평점(또는 점수)을 예측
유저-아이템 행렬을 채우는 문제
유저(m명)-아이템(n개)의 m X n 행렬, 그러나 비어 있는 부분 존재
	-관측값(Observed value)은 모델 학습에 사용(어떤 유저가 어떤 평점을 줬는지)
	-결측값(Missing value)은 모델 평가에 사용(유저가 평가하지 않은 부분)
```

## 03. Ch 01. 추천시스템이란 - 03. 추천시스템의 알고리즘 종류

```python
<추천 시스템과 추천 알고리즘>
Item Pool -> Generate Candidates -> predict Intentions or Filtering -> Ranking

<Contents-based Recommender System 컨텐츠 기반 추천 시스템>
컨텐츠가 얼마나 유사한지를 기반으로 컨텐츠 추천
사용자가 과거에 좋아했던 아이템을 파악하고, 그 아이템과 비슷한 아이템을 추천
1.유저가 과거에 접한 아이템이면서 만족한 아이템
2.유저가 좋아했던 아이템 중 일부 또는 전체와 비슷한 아이템 선정
3.선정된 아이템을 유저에게 추천
예. 스파이더맨에 4.5점 평점 부여한 유저 -> 타이타닉보다 캡틴 마블을 더 좋아할 것이다.

<Collaborative Filtering 협업 필터링>
비슷한 성향 또는 취향을 갖는 다른 유저가 좋아한 아이템을 현재 유저에게 추천
간단하면서 수준 높은 정확도를 나타냄
1.유저 A와 B 모두 같은 아이템에 대해 비슷한 또는 같은 평가를 했다
2.이 때 유저 A는 다른 아이템에도 비슷한 호감을 나타냈다
3.유저 A, B의 성향은 비슷할 것이므로 다른 아이템을 유저 B에게도 추천
예. 스파이더맨에 4.5점을 준 2명의 유저 -> 유저 A가 과거 좋아했던 캡틴 마블을 유저 B에게 추천

<Hybrid Recommender System>
Content-based와 Collaborative Filtering의 장단점을 상호보완
Collaborative Filtering은 새로운 아이템에 대한 추천 부족
content-based 기법이 cold-start 문제에 도움을 줄 수 있음

내가 좋아하는 컨텐츠와 비슷한 컨텐츠를 보여주는 게 컨텐츠 기반 추천 시스템, 다른 사람 것을 보여주는 게 협업 필터링

<다른 추천 알고리즘>
1.Context-based Recommendation
context-aware Recommendation system
location-based Recommendation system
real-time or time-sensitive Recommendation system

2.Community-based Recommendation
사용자의 친구 또는 속한 커뮤니티의 선호도를 바탕으로 추천
SNS 등의 뉴스피드 또는 SNS 네트워크 데이터 등 활용

3.Knowledge-based Recommendation
특정 도메인 지식을 바탕으로 아이템의 피처를 활용한 추천
Case-based Recommendation
	-사용자의 니즈(현재 문제 등)와 해결책 중 가장 적합한 것을 골라서 추천
Constraint-based Recommendation
	-사용자에게 추천할 때, 정해진 규칙을 바탕으로 추천
	-스코어링을 통해 리스트가 나왔다 하더라도 정해진 규칙이 존재한다면 규칙 적용된 리스트가 최종 출력
```

## 04. Ch 01. 추천시스템이란 - 04. 추천시스템의 한계

```python
<퀵 리뷰>
1.추천시스템은 프로파일링한 유저와 아이템으로 적절한 것을 추천
2.추천시스템은 유저와 아이템의 관계를 파악하고 점수화한다
3.추천시스템은 푸시 인포메이션, 검색 서비스는 풀 인포메이션
4.추천시스템은 랭킹 문제 또는 예측 문제이다.
5.컨텐츠기반, 협업필터링, 딥러닝 기반 등 추천 알고리즘 종류는 다양하다

넷플릭스 콘텐츠 추천, 인스타그램 팔로우 추천, 쿠팡 상품 추천

<추천 시스템 한계1>
-Scalability(확장성(사용자 수의 증대에 유연하게 대응할 수 있는 정도))
실제 서비스 상황은 다양한 종류의 데이터
학습 또는 분석에 사용한 데이터와는 전혀 다른 실전 데이터

-Proactive Recommender System
특별한 요청이 없어도 사전에 먼저 제공하는 추천 서비스
모바일, 인터넷 등 어디서는 유저에게 끊임없이 좋은 정보를 추천할 수 있는 서비스

-Cold-start Problem
추천서비스를 위한 데이터 부족
기본적인 성능을 보장하는 협업필터링 모델 구축이 쉽지 않은 것이 일반적
컨텐츠 기반 또는 지식 기반의 방법 역시 서비스로 적용하기 어려움

-Privacy preserving Recommender system
개인정보 등 유저 정보가 가장 중요하지만 직접적으로 사용하기 어려움
아이템 유저 프로파일링 하는 것이 가장 중요하지만 개인정보는 굉장히 민감
이걸 어떻게 직접적으로 사용할 수 있을지를 고민해봐야
Implicit 정보(로그 등)를 유저 정보로 많이 택하고 있는 현실, 여전히 개인정보 이슈… 직접 사용 어려워

<추천 시스템 한계2>
-Mobile devices and Usage Contexts
Personal Computing, Location-based Service(LBS)
개별 상황 또는 환경 등에 따라 다른 컨텍스트를 사용

-Long-term and Short-term use preference
개인 또는 그룹의 단기 장기 관심사항
추천받고 싶은 아이템이 현재 또는 과거 중 어느 시기와 관련 있는지 파악하기 어려움

-Generic User models and Cross Domain Recommender System
하나의 모델을 여러가지 데이터에 적용하기 어려움
비슷한 도메인의 데이터를 활용해도 동일한 성능의 추천시스템을 기대하기 어려움
예. 한국영화, 미국 영화: 데이터 자체가 다름

-Starvation and Diversity
Starvation: 필요한 컴퓨터 자원을 끊임없이 가져오지 못하는 상황
유저 아이템이 다양하고 모든 유저 아이템에 더 많은 관심을 부여해야 함

<Long Tail Economy>
양쪽 사이드를 모두 고려. 롱테일 이코노미에 최적화된 알고리즘 만들기 위해 고민해야.
<그리하여 추천 시스템은 정답이 없습니다!>
정량화된 평가 어렵지만 구매 증가, 방문 횟수 등을 통해 간접적으로 알 수 있다.
```
