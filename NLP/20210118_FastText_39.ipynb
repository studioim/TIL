```python
!head -n 2 ./ko.tok.vec
```

```python
# 9348개의 워드 임베딩 벡터, 각각은 256차원이라는 얘기
# 굉장히 dense 하고 0이 없다.
# TF 같은 경우에는 정규화(normalization)의 이슈가 있을 수 있지만 fasttext는 비교적 0 근처에 잘 몰려 있는 것을 확인할 수 있다.
# 성능 개선에 좋은 영향을 미칠 수 있다.
# fasttext 쉽게 가장(?) 빠르게 워드 임베딩 할 수 있다.
# 그렇다고 NLP를 하는 데 크게 쓸모가 있지는 않음.
# 뉴럴 네트워크는 엔드투엔드를 지향하기 때문에 워드 임베딩 벡터가 최종 결과물이 돼야 하는 태스크는 거의 없다.
# 이와 같은 워드 임베딩 벡터를 구해서 뉴럴 네트워크에 집어넣는 것이 아니고
# 파이토치 내에 임베딩 레이어라는 것을 활용해서 그곳에 원핫 벡터를 통과시키면 태스크에 알맞는 워드임베딩 벡터를 자동으로 변환해준다.
```
