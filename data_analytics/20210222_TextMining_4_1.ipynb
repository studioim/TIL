# 제4장 비정형 데이터 마이닝

## 제1절 텍스트 마이닝
- 텍스트 마이닝: 텍스트를 사용해 패턴이나 관계를 추출하고 그 안에서 의미 있는 정보나 가치를 발굴해 해석하는 일련의 과정을 통칭
    - 다양한 포맷의 문서로부터 데이터를 획득해 이를 문서별 단어의 매트릭스로 만들어 추가 분석이나 데이터 마이닝 기법을 적용해 통찰을 얻거나 의사결정을 지원하는 방법
    - 설비 고장 예측, 상품평 분석, 뉴스기사나 연설문의 시간에 따른 주제의 흐름 파악, 특정 분야의 전문가를 알아내는 데 활용

### 1. 텍스트 마이닝의 기능
- 문서 요약
- 문서 분류
- 문서 군집
- 특성 추출

### 2. 텍스트 마이닝의 기본 프로세스
- 가. 데이터 수집
    - 웹페이지에서 HTML을 데이터로 가져와서 파싱하는 방식
    - API를 이용해 가져오는 방식
- 나. 데이터 전처리 및 가공
    - 1) 코퍼스 생성
        - 코퍼스는 데이터 마이닝의 절차 중 데이터의 정제, 통합, 선택, 변환의 과정을 거친 구조화된 단계
        - 더 이상 추가적 절차 없이 데이터 마이닝 알고리즘 실험에서 활용될 수 있는 상태를 의미
    - 2) 데이터 전처리
        - 빈 공간(화이트 스페이스) 제거
        - 특정 패턴을 가지는 단어를 제거하거나 다른 단어로 대체
        - 문장 부호 및 구두점 제거
        - 대문자 소문자로 변경
            - 한글이나 특수문자가 있는 경우 오류를 발생시키므로 한글이나 특수문자를 미리 처리한 뒤 진행
        - 특정단어 제거(불용어)
            - 접속사, 대명사, 한국어 조사 등
- 다. 자연어 처리
    - 텍스트 마이닝 시 자연어 처리는 기본적으로 형태소 분석을 하는 과정을 포함
    - 불용어를 처리하는 과정에서 접속사 대명사 등을 제거
    - 공통 어간을 가지는 단어를 묶기 위해 스테밍(stemming) 처리
- 라. TDM(Term-Document Matrix) 구축
    - 앞선 과정을 통해 문서의 빈 공간을 제거하고, 대문자를 소문자로 변환, 문장부호 제거, 불용어(제외어) 처리, 스테밍 등의 과정을 수행
    - 이렇게 전처리된 데이터에서 각 문서와 단어 간의 사용 여부를 이용해 만들어진 매트릭스(행렬)가 바로 TDM
    - TDM을 보면 등장한 각 단어의 빈도수를 쉽게 알 수 있다
    - 1) 모든 단어
        - 함수를 실행할 때 딕셔너리 부분을 작성하지 않으면 분리된 모든 단어가 Term으로써 TDM을 만드는 데 사용
    - 2) 단어 사전(딕셔너리)
        - 복수의 문자들의 집합으로 텍스트 마이닝에서 분석에 사용하고자 하는 단어들의 집합
- 마. 분석 및 시각화
    - 1) Association
        - 연관 분석(association analysis): 어떤 사건이 동시에 일어나는지를 알아보는 것
        - apriori 알고리즘에서 연관분석 수행 위해 빈발 아이템 집합과 연관 규칙이라고 하는 두 가지 형태로 관계를 표현
        - 아이템의 빈발 정도를 측정하기 위해 지지도(서포트)의 개념을, 연관 규칙 관계의 강도를 측정하기 위해 신뢰도(컨피던스) 개념을 사용
    - 2) 워드 클라우드
        - 문서에 포함되는 단어의 사용 빈도를 효과적으로 보여주기 위해 사용
        - TDM 이용
    - 3) 감성 분석(sentiment analysis)
        - 오피니언 마이닝 등으로 언급
        - 문장에서 사용된 단어의 긍정 부정 여부에 따라 얼마나 긍정적인 단어가 많은지를 보고 특정 소스를 부여해 어떤 색을 띄고 있는 문장인지를 평가
        - 각 문장이 긍정인지 부정인지는 주체에 따라 다르게 해석할 수 있음
        - 개별 문장의 분석에서 오류가 생기더라도 수많은 문서나 데이터를 가공하면 그 추이를 파악하는 데는 큰 무리가 없음
            - 긍정 부정 단어 파일을 읽어와 저장
            - 문장부호, 특수 기호, 숫자 문자열을 빈칸으로 대체
            - 대문자를 소문자로 변환
            - 문장을 단어 리스트로 변환
            - 문장에서 긍부정 단어가 나타날 때마다 해당 위치를 1로 세팅해 저장
            - 긍정과 부정 단어의 합을 차감해 점수 산출
            - 점수와 문장을 데이터 프레임 형태로 반환
                - 단어들이 붙어 있는 것들은 구분하지 못해 0점 처리, 긍정 1점, 부정 -1점 부여
